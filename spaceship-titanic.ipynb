```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier as RFC
```

```python
train_data = pd.read_csv("/kaggle/input/competitions/spaceship-titanic/train.csv")
train_data.head()

test_data = pd.read_csv("/kaggle/input/competitions/spaceship-titanic/test.csv")
```

```python
train_data.describe()
X = train_data.drop(["Transported", "PassengerId"], axis = 1)
y = train_data["Transported"]
X_test = test_data.drop("PassengerId", axis = 1)
```

```python
train_data.isnull().sum()
```

```python
# find categorical null 
cat_col = X.select_dtypes(include="object").columns
for col in cat_col:
    X[col] = X[col].fillna("None")
    X_test[col] = X_test[col].fillna("None")
```

```python
# find numerical null 
num_col = X.select_dtypes(exclude="object").columns
for col in num_col:
    median_val = X[col].median()
    X[col] = X[col].fillna(median_val)
    X_test[col] = X_test[col].fillna(median_val)
```

```python
X_train = pd.get_dummies(X)
X_test = pd.get_dummies(test_data)

rfc = RFC(n_estimators=100,max_depth=5, random_state=1)

X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

rfc.fit(X_train,y)
predictions = rfc.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Transported': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")







```
